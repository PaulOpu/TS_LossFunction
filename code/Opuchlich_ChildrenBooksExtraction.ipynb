```python
import pandas as pd
import re
import numpy as np
```

```python
with open("data/CBTest_childbook/data/cbt_train.txt","r") as f:
    data = f.read().split("\n")
```

```python
regex_chapter_start = re.compile(r"^[A-Z]{2,2}")
```

```python
starting_lines = [ind for ind,line in enumerate(data) if regex_chapter_start.findall(line)]
```

```python
groups = (np.array(starting_lines)/100).astype(int)
```

```python
prev_group = 0
line_groups = []
for ind in range(len(groups)):
    if prev_group != groups[ind]:
        line_groups += [starting_lines[ind-1]]
        prev_group = groups[ind]
```
